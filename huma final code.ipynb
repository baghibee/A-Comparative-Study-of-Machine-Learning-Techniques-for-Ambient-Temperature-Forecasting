```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import logging
from sklearn.svm import SVR
from sklearn.linear_model import Ridge, Lasso
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python


# Define the path
data_path = '/content/drive/MyDrive/weatherStation/daily'



# Set up logging
logging.basicConfig(level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s')

# Define column name configurations
column_configurations = {
    'config1': [
        'Time', 'Wind Dir', 'Wind Spd', 'Wind Gust', 'Hum In', 'Humidity', 'Temp In', 'Temp', 'Raw Barom',
        'UV', 'Solar', 'Dew Point', 'DailyRain', 'RainRate'
    ],
    'config2': [
        'Time', 'Wind Dir', 'Wind Spd', 'Hum In', 'Humidity', 'Temp In', 'Temp', 'Raw Barom', 'Temp Ch 1',
        'Hum Ch 1', 'Temp Ch 2', 'Hum Ch 2', 'UV', 'Solar', 'Dew Point', 'RainRate'
    ],
    'config3': [
        'Time', 'Wind Dir', 'Wind Spd', 'Wind Gust', 'Hum In', 'Humidity', 'Temp In', 'Temp', 'Raw Barom',
        'UV', 'Solar', 'Dew Point', 'DailyRain', 'RainRate'
    ],
    'config4': [
        'Time', 'Wind Dir', 'Wind Spd', 'Hum In', 'Humidity', 'Temp In', 'Temp', 'Raw Barom', 'UV', 'Solar',
        'Dew Point', 'DailyRain', 'RainRate'
    ]
}

# Initialize an empty list to store all dataframes
dfs = []

# Iterate through each file in the directory
for filename in os.listdir(data_path):
    if filename.endswith(".txt"):
        file_path = os.path.join(data_path, filename)

        try:
            # Read the first line to get the date
            with open(file_path, 'r', encoding='latin1') as f:
                first_line = f.readline().strip()
                # Skip the second line containing units
                f.readline()

            # Extract the date from the first line
            try:
                file_date_from_line = pd.to_datetime(first_line, format='%d/%m/%y').date()
            except ValueError:
                logging.warning(f"Date format in file '{filename}' is not recognized. Skipping.")
                continue

            # Read the remaining data into a DataFrame with a placeholder for column names
            df = pd.read_csv(file_path, delimiter='\s+', skiprows=3, encoding='latin1', header=None, on_bad_lines='skip')

            # Check which column configuration matches the DataFrame
            for config_name, columns in column_configurations.items():
                if df.shape[1] == len(columns):
                    df.columns = columns
                    df['Date'] = file_date_from_line
                    dfs.append(df)
                    logging.info(f"File '{filename}' processed with column configuration '{config_name}'.")
                    break
            else:
                logging.warning(f"File '{filename}' does not match any known column configuration. Skipping.")

        except (UnicodeDecodeError, pd.errors.EmptyDataError, pd.errors.ParserError, ValueError) as e:
            logging.error(f"Could not read {file_path}: {e}")

# Concatenate all dataframes into a single dataframe, ignoring empty or all-NA entries
if dfs:
    weather_data = pd.concat(dfs, ignore_index=True)
    logging.info(f"Successfully concatenated data into a single DataFrame with {weather_data.shape[0]} rows.")
else:
    logging.warning("No dataframes to concatenate.")
```

## Convert Object Values into Numerical Float values

```python
import re

# List of columns to convert to numeric types
columns_to_convert = [ 'Wind Dir', 'Wind Spd', 'Wind Gust', 'Hum In', 'Humidity',
       'Temp In', 'Temp', 'Raw Barom', 'UV', 'Solar', 'Dew Point', 'DailyRain',
       'RainRate', 'Temp Ch 1', 'Hum Ch 1', 'Temp Ch 2', 'Hum Ch 2']

def clean_column(column):
    # Remove any non-numeric characters (except for '.' and '-')
    return column.replace(regex=True, to_replace=r'[^0-9.-]', value='')

# Apply the cleaning function to each column
for col in columns_to_convert:
    weather_data[col] = clean_column(weather_data[col])

# Convert to numeric after cleaning
for col in columns_to_convert:
    weather_data[col] = pd.to_numeric(weather_data[col], errors='coerce')
```

## Convert 'Date' and 'Time' to a single 'Datetime' column

```python
# Convert 'Date' and 'Time' to a single 'Datetime' column
weather_data['Datetime'] = pd.to_datetime(
    weather_data['Date'].astype(str) + ' ' + weather_data['Time'],
    format='%Y-%m-%d %H:%M',
    errors='coerce'  # Converts invalid parsing to NaT
)
```

## Removing Missing values and Visualizing it

```python
# Remove rows with missing values
daily_weather_data_cleaned = daily_weather_data.dropna()

# Visualize the cleaned data for 'Temp'
plt.figure(figsize=(15, 6))
plt.plot(daily_weather_data_cleaned['Datetime'], daily_weather_data_cleaned['Temp'], label='Temperature (°C)', color='black')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.title('Daily Average Temperature (Cleaned Data)')
plt.legend(loc='upper left')
plt.grid(True)
plt.show()
```

# Train, Validation and Test Split
### Features are scaled and split into training, testing and validation

```python
# Assuming 'Temp' is the target variable and the rest are features
X = daily_weather_data_cleaned.drop(['Temp', 'Datetime'], axis=1)
y = daily_weather_data_cleaned['Temp']

# Scale the features
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Split into train, validation, and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("Train set size:", X_train.shape, y_train.shape)
print("Test set size:", X_test.shape, y_test.shape)
```

# SVM Model

```python
# Define the parameter grid
svm_param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf', 'linear']
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(SVR(), svm_param_grid, refit=True, verbose=2, cv=5)
grid_search.fit(X_train, y_train)


# Print the best parameters
print("Best parameters found: ", grid_search.best_params_)
```

```python
# Evaluate the best model
best_svm_model = grid_search.best_estimator_
y_pred_best_svm = best_svm_model.predict(X_test)
```

```python
# Visualize predictions vs actual values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_best_svm)
plt.xlabel('Actual Temperature')
plt.ylabel('Predicted Temperature (RBF)')
plt.title('Actual vs Predicted Temperature (RBF Kernel)')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], linestyle='--', color='red')
plt.grid(True)
plt.show()
```

```python
# Calculate the metrics
me_svm = np.mean(y_pred_best_svm - y_test)
mae_svm = mean_absolute_error(y_test, y_pred_best_svm)
rmse_svm = np.sqrt(mean_squared_error(y_test, y_pred_best_svm))
r2_svm = r2_score(y_test, y_pred_best_svm)

print("Mean Error (ME):", me_svm)
print("Mean Absolute Error (MAE):", mae_svm)
print("Root Mean Squared Error (RMSE):", rmse_svm)
print("R-squared (R2):", r2_svm)
```

```python
# Ensure the Datetime column is in datetime format
daily_weather_data_cleaned['Datetime'] = pd.to_datetime(daily_weather_data_cleaned['Datetime'])

# Create a DataFrame to hold the predictions and the true values
results = pd.DataFrame({
    'Datetime': daily_weather_data_cleaned.loc[y_test.index, 'Datetime'],
    'True Temp': y_test,
    'Predicted Temp': y_pred_best_svm
})

# Sort by datetime
results = results.sort_values(by='Datetime')

# Plot the true vs predicted values
plt.figure(figsize=(14, 7))
plt.plot(results['Datetime'], results['True Temp'], label='True Temp', color='black')
plt.plot(results['Datetime'], results['Predicted Temp'], label='Predicted Temp', color='red', linestyle='dashed')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('True vs Predicted Temperature')
plt.legend()
plt.show()
```

# MLR

```python
# Define the parameter grid for Ridge Regression
mlr_param_grid = {
    'alpha': [0.1, 0.01, 0.5, 1, 10, 100]
}

# Perform grid search with cross-validation for Ridge Regression
grid_search_ridge = GridSearchCV(Ridge(), mlr_param_grid, refit=True, verbose=2, cv=5)
grid_search_ridge.fit(X_train, y_train)

# Print the best parameters for Ridge
print("Best parameters found for Ridge: ", grid_search_ridge.best_params_)

# Evaluate the best Ridge model
best_ridge = grid_search_ridge.best_estimator_
y_pred_ridge = best_ridge.predict(X_test)
```

```python
# Perform grid search with cross-validation for Lasso Regression
grid_search_lasso = GridSearchCV(Lasso(), mlr_param_grid, refit=True, verbose=2, cv=5)
grid_search_lasso.fit(X_train, y_train)

# Print the best parameters for Lasso
print("Best parameters found for Lasso: ", grid_search_lasso.best_params_)

# Evaluate the best Lasso model
best_lasso = grid_search_lasso.best_estimator_
y_pred_lasso = best_lasso.predict(X_test)
```

```python
# Calculate the metrics for Ridge
mae_ridge = mean_absolute_error(y_test, y_pred_ridge)
rmse_ridge = np.sqrt(mean_squared_error(y_test, y_pred_ridge))
r2_ridge = r2_score(y_test, y_pred_ridge)

print("Ridge Regression Mean Absolute Error (MAE):", mae_ridge)
print("Ridge Regression Root Mean Squared Error (RMSE):", rmse_ridge)
print("Ridge Regression R-squared (R2):", r2_ridge)
```

```python
# Calculate the metrics for Lasso
mae_lasso = mean_absolute_error(y_test, y_pred_lasso)
rmse_lasso = np.sqrt(mean_squared_error(y_test, y_pred_lasso))
r2_lasso = r2_score(y_test, y_pred_lasso)

print("Lasso Regression Mean Absolute Error (MAE):", mae_lasso)
print("Lasso Regression Root Mean Squared Error (RMSE):", rmse_lasso)
print("Lasso Regression R-squared (R2):", r2_lasso)
```

```python
# Ensure the Datetime column is in datetime format
daily_weather_data_cleaned['Datetime'] = pd.to_datetime(daily_weather_data_cleaned['Datetime'])

# Create DataFrames to hold the predictions and the true values
results_ridge = pd.DataFrame({
    'Datetime': daily_weather_data_cleaned.loc[y_test.index, 'Datetime'],
    'True Temp': y_test,
    'Predicted Temp': y_pred_ridge
})

results_lasso = pd.DataFrame({
    'Datetime': daily_weather_data_cleaned.loc[y_test.index, 'Datetime'],
    'True Temp': y_test,
    'Predicted Temp': y_pred_lasso
})

# Sort by datetime
results_ridge = results_ridge.sort_values(by='Datetime')
results_lasso = results_lasso.sort_values(by='Datetime')

# Plot the true vs predicted values for Ridge
plt.figure(figsize=(14, 7))
plt.plot(results_ridge['Datetime'], results_ridge['True Temp'], label='True Temp', color='black')
plt.plot(results_ridge['Datetime'], results_ridge['Predicted Temp'], label='Predicted Temp (Ridge)', color='red', linestyle='dashed')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('True vs Predicted Temperature (Ridge Regression)')
plt.legend()
plt.show()
```

```python
# Plot the true vs predicted values for Lasso
plt.figure(figsize=(14, 7))
plt.plot(results_lasso['Datetime'], results_lasso['True Temp'], label='True Temp', color='black')
plt.plot(results_lasso['Datetime'], results_lasso['Predicted Temp'], label='Predicted Temp (Lasso)', color='red', linestyle='dashed')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('True vs Predicted Temperature (Lasso Regression)')
plt.legend()
plt.show()
```

# Comparision

```python
# Define the metrics for each model
metrics = {
    'Lasso': {
        'MAE': mae_lasso,
        'RMSE': rmse_lasso,
        'R2': r2_lasso
    },
    'Ridge': {
        'MAE': mae_ridge,
        'RMSE': rmse_ridge,
        'R2': r2_ridge
    },
    'SVM': {
        'MAE': mae_svm,
        'RMSE': rmse_svm,
        'R2': r2_svm
    }
}

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Define bar width and positions
bar_width = 0.2
models = list(metrics.keys())
metrics_names = list(metrics[models[0]].keys())
num_metrics = len(metrics_names)
index = np.arange(len(models))

# Plot each metric
for i, metric_name in enumerate(metrics_names):
    values_list = [metrics[model][metric_name] for model in models]
    bar_positions = index + i * bar_width
    bars = ax.bar(bar_positions, values_list, bar_width, label=metric_name)

# Set labels and title
ax.set_xlabel('Model')
ax.set_ylabel('Scores')
ax.set_title('Comparison of Regression Models')
ax.set_xticks(index + bar_width * (num_metrics / 2 - 0.5))
ax.set_xticklabels(models)
ax.legend(title='Metrics')

# Show the plot
plt.tight_layout()
plt.show()
```
